```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulated outputs from models (replace if needed)

eczema_preds = np.random.randint(0, 2, 20)
stress_preds = np.random.randint(0, 2, 20)
scratch_preds = np.random.randint(0, 2, 20)
```

```python
def final_prediction(eczema, stress, scratch):
    if eczema == 1 and stress == 1 and scratch == 1:
        return 2   # High
    elif eczema == 1 and (stress == 1 or scratch == 1):
        return 1   # Moderate
    else:
        return 0   # Low
```

```python
final_preds = []

for i in range(len(eczema_preds)):
    result = final_prediction(
        eczema_preds[i],
        stress_preds[i],
        scratch_preds[i]
    )
    final_preds.append(result)

final_preds = np.array(final_preds)
```

```python
for i in range(5):
    print(f"Sample {i}:")
    print("Eczema:", eczema_preds[i],
          "Stress:", stress_preds[i],
          "Scratch:", scratch_preds[i],
          "→ Final:", final_preds[i])
    print()
```

```python
labels = ["Low", "Moderate", "High"]
counts = [
    np.sum(final_preds == 0),
    np.sum(final_preds == 1),
    np.sum(final_preds == 2)
]

plt.bar(labels, counts)
plt.title("Flare Risk Distribution")
plt.xlabel("Risk Level")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()

plt.plot(eczema_preds, label="Eczema", marker='o')
plt.plot(stress_preds, label="Stress", marker='x')
plt.plot(scratch_preds, label="Scratch", marker='s')

plt.title("Model Outputs Comparison")
plt.xlabel("Sample Index")
plt.ylabel("Prediction")
plt.legend()

plt.show()
```

```python
plt.figure()

plt.plot(final_preds, label="Final Risk", linewidth=3)

plt.title("Final Flare Risk Prediction")
plt.xlabel("Sample Index")
plt.ylabel("Risk Level (0=Low,1=Moderate,2=High)")
plt.legend()

plt.show()
```
